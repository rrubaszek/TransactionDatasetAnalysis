```python
%load_ext autoreload
%autoreload 2

from transaction_analysis.data import io
from transaction_analysis.data.preprocess import (
    currency_to_int32,
    int64_downcast,
    month_year_to_datetime,
    str_to_category,
    yes_no_to_bool,
)
from transaction_analysis.paths import FRAUD_DATASET_DIR

in_file = FRAUD_DATASET_DIR / "raw" / "cards_data.csv"
out_file = FRAUD_DATASET_DIR / "preprocessed" / "cards.parquet"
```

```python
df = io.read_csv(in_file)
df.info()
```

```python
df = int64_downcast(df, "id")
```

```python
df = int64_downcast(df, "client_id")
```

```python
df = str_to_category(df, "card_brand")
```

```python
df = str_to_category(df, "card_type")
```

```python
df = month_year_to_datetime(df, "expires")
```

```python
df = int64_downcast(df, "cvv")
```

```python
df = yes_no_to_bool(df, "has_chip")
```

```python
df = int64_downcast(df, "num_cards_issued")
```

```python
df = currency_to_int32(df, "credit_limit")
```

```python
df = month_year_to_datetime(df, "acct_open_date")
```

```python
df = int64_downcast(df, "year_pin_last_changed")
```

```python
df = yes_no_to_bool(df, "card_on_dark_web")
```

```python
df = df.set_index("id")
df.info()
```

```python
io.to_parquet(df, out_file)
```
