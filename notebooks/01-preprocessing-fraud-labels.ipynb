```python
%load_ext autoreload
%autoreload 2

from transaction_analysis.data import io
from transaction_analysis.data.preprocess import int64_downcast, yes_no_to_bool
from transaction_analysis.paths import FRAUD_DATASET_DIR

in_file = FRAUD_DATASET_DIR / "raw" / "train_fraud_labels.json"
out_file = FRAUD_DATASET_DIR / "preprocessed" / "fraud_labels.parquet"
```

```python
df = io.read_json(in_file)
df.head()
```

```python
df = df.rename_axis("id")
```

```python
df = df.rename(columns={"target": "fraud"})
```

```python
df = df.reset_index()
```

```python
df = int64_downcast(df, "id")
```

```python
df = yes_no_to_bool(df, "fraud")
```

```python
df = df.set_index("id")
df.info()
```

```python
io.to_parquet(df, out_file)
```
