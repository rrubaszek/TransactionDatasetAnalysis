```python
%load_ext autoreload
%autoreload 2

from transaction_analysis.data import io
from transaction_analysis.data.preprocess import (
    currency_to_int32,
    int64_downcast,
    str_to_category,
    str_to_datetime,
    zip_to_category,
)
from transaction_analysis.paths import FRAUD_DATASET_DIR

in_file = FRAUD_DATASET_DIR / "raw" / "transactions_data.csv"
out_file = FRAUD_DATASET_DIR / "preprocessed" / "transactions.parquet"
```

```python
df = io.read_csv(in_file)
df.info()
```

```python
df = int64_downcast(df, "id")
```

```python
df = str_to_datetime(df, "date")
```

```python
df = int64_downcast(df, "client_id")
```

```python
df = int64_downcast(df, "card_id")
```

```python
df = currency_to_int32(df, "amount")
```

```python
df = str_to_category(df, "use_chip")
```

```python
df = int64_downcast(df, "merchant_id")
```

```python
df = str_to_category(df, "merchant_city")
```

```python
df = str_to_category(df, "merchant_state")
```

```python
df = zip_to_category(df, "zip")
```

```python
df = int64_downcast(df, "mcc")
```

```python
df = str_to_category(df, "errors")
```

```python
df = df.set_index("id")
df.info()
```

```python
io.to_parquet(df, out_file)
```
