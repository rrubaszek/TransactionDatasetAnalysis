```python
%load_ext autoreload
%autoreload 2

from transaction_analysis.data import io
from transaction_analysis.data.preprocess import (
    currency_to_int32,
    int64_downcast,
    str_to_category,
)
from transaction_analysis.paths import FRAUD_DATASET_DIR

in_file = FRAUD_DATASET_DIR / "raw" / "users_data.csv"
out_file = FRAUD_DATASET_DIR / "preprocessed" / "users.parquet"
```

```python
df = str_to_category(df, "gender")
```
