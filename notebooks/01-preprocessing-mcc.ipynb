```python
%load_ext autoreload
%autoreload 2

from transaction_analysis.data import io
from transaction_analysis.data.preprocess import int64_downcast
from transaction_analysis.paths import FRAUD_DATASET_DIR

in_file = FRAUD_DATASET_DIR / "raw" / "mcc_codes.json"
out_file = FRAUD_DATASET_DIR / "preprocessed" / "mcc_codes.parquet"
```

```python
s = io.read_json(in_file, typ="series")
s.head()
```

```python
s = s.rename_axis("id")
```

```python
s = s.rename("description")
```

```python
df = s.reset_index()
```

```python
df = int64_downcast(df, "id")
```

```python
df = df.set_index("id")
df.info()
```

```python
io.to_parquet(df, out_file)
```
